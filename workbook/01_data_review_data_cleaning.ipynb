As there are several so many subcategories, I decided to create a new columns having a more aggregated view. 

```python
df["service_code_summary"] = df["SERVICE CODE"]
```

```python
# STRAY: STRAY and STRAY AN SECURED
df.loc[(df["SERVICE CODE"] == "STRAY") | (df["SERVICE CODE"] == "STRAY AND SECURED"),["service_code_summary"]] = "STRAY"
```

```python
# BITE CASE: 
df.loc[(df["SERVICE CODE"] == "BITE CASE")  | (df["SERVICE CODE"] == "BITE CASE SECURE")|(df["SERVICE CODE"] == "BITE CASE LOOSE"),["service_code_summary"]] = "BITE CASE"
```

```python
# PUT "HIT BY CAR" to "INJURED"
df.loc[(df["SERVICE CODE"] == "HIT BY CAR"), ["service_code_summary"]] = "INJURED"
```

```python
# COMBINE"VET PICKUP" AND "SICK ANIMAL"
df.loc[(df["SERVICE CODE"] == "VET PICKUP"), ["service_code_summary"]] = "SICK ANIMAL"
```

```python
# COMBINE "LOOSE LIVESTOCK" TO "LOOSE"
df.loc[(df["SERVICE CODE"] == "LOOSE LIVESTOCK"), ["service_code_summary"]] = "LOOSE"
```

```python
# ADD "SNAKE" TO "DANGEROUS ANIMAL SIT"
df.loc[(df["SERVICE CODE"] == "SNAKE"), ["service_code_summary"]] = "DANGEROUS ANIMAL SIT"
```

```python
# COMBINE "SPECIAL CASE", "UNKNOW", "SCHOOL CALL" to "OTHER"
df.loc[(df["SERVICE CODE"] == "SPECIAL CASE"), ["service_code_summary"]] = "OTHER"
df.loc[(df["SERVICE CODE"] == "UNKNOWN"), ["service_code_summary"]] = "OTHER"
df.loc[(df["SERVICE CODE"] == "SCHOOL CALL"), ["service_code_summary"]] = "OTHER"
df.loc[(df["SERVICE CODE"] == "OWNED"), ["service_code_summary"]] = "OTHER"
```
